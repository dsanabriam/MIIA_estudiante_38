### Ejercicio 1
Utiliza `plotly` para realizar un  diagramas de torta sobre el estado conyugal de los clientes (resaltando el más predominante).

```python
# YOUR CODE HERE
EstadoCivil= df[["Marital_Status", "CLIENTNUM"]].groupby(["Marital_Status"]).agg("count")
labels = sorted(df["Marital_Status"].unique().tolist())
values = [i for i in EstadoCivil["CLIENTNUM"]]

fig = go.Figure(data=[go.Pie(labels=labels, values=values,textinfo='label+percent', pull=[0, 0.2, 0, 0])])
fig.update_layout(title_text='Estado Conyugal de los Clientes')
fig.show()
```

### Ejercicio 2
Utiliza `plotly` para realizar un gráfico que permita verificar si el cambio en el monto de transacciones (cuarto trimestre a primer trimestre) se ajusta a una distribución normal.

```python
# YOUR CODE HERE
import plotly.figure_factory as gfn

x= df["Total_Ct_Chng_Q4_Q1"]

group_labels = ["Total_Ct_Chng_Q4_Q1"]
fig = gfn.create_distplot([x], group_labels, bin_size=.01,
                         curve_type='normal' 
                         )
fig.update_layout(title_text='Cambio para el trimestre Q4 vs Q1 en el monto de las transacciones vs Normal Distribution')
fig.show()
```

### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado.
2. Número de niños en el hogar.
3. Número de adolescentes en el hogar.

```python
# YOUR CODE HERE
def valores_filtro_columna(columna):
    columna = columna.astype(str)       # Convertimos en str los valores de la columna que entra por parámetro.
    valores = columna.unique().tolist() # Guardamos una lista con los valores únicos de la columna que entra por parámetro.
    valores = sorted(valores)           # Organizamos la lista de valores únicos de manera ascedente.
    valores.insert(0, "Todos")          # Agregamos el elemento "Todos" en la primera posición de la lista.
    return valores 


def filtrar(max_escolaridad, ninos_hogar, teen_hogar):
    df = mercadeo
    if  max_escolaridad != "Todos":
        display(df[(df["Education"] == max_escolaridad) & (df["Kidhome"] <= ninos_hogar) & (df["Teenhome"] <= teen_hogar) ])
    else:
        display(df[(df["Kidhome"] <= ninos_hogar)&(df["Teenhome"] <= teen_hogar)])
        
Cuadro_Interactivo = interactive(
                                filtrar, 
                                max_escolaridad = valores_filtro_columna(mercadeo["Education"]),
                                ninos_hogar = (mercadeo["Kidhome"].min(), mercadeo["Kidhome"].max(), 1)  ,
                                teen_hogar = (mercadeo["Teenhome"].min(), mercadeo["Teenhome"].max(), 1)
                              )

display(Cuadro_Interactivo)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
# YOUR CODE HERE
df = mercadeo[['Income', 'MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts', 'MntGoldProds']]
df_unpvt = df.melt(id_vars=['Income'], var_name = 'tipo_gasto', value_name = 'gasto')


tipo = widgets.Dropdown(
                        description = 'Tipo de gasto:',
                        options = df_unpvt['tipo_gasto'].unique().tolist(),
                        style = {'description_width': 'initial'}       
                       )


f1 = go.FigureWidget(
                        data = go.Scatter(name = 'Gastos - Ingresos anuales del hogar del cliente', mode='markers', marker_color='rgba(152, 0, 0, .8)'),  
    
                        layout = go.Layout(
                                            
                                            xaxis_title = 'Ingresos anuales del hogar del cliente',
                                            yaxis_title = 'Gastos',
                                          )
                     )


f2 = go.FigureWidget(
                        data = go.Box(name = 'Gastos'),  
    
                        layout = go.Layout(
                                            
                                            yaxis_title = 'Gastos',
                                            
                                          )
                     )


def filtrar():
    Lista_Filtros = [i for i in df_unpvt['tipo_gasto'] == tipo.value]
    df_filtros = df_unpvt[Lista_Filtros]
    f1.data[0].x = df_filtros['Income']
    f1.data[0].y = df_filtros['gasto'] 
    f2.data[0].y = df_filtros['gasto'] 
    f1.update_layout(title_text= 'Ingresos anuales del hogar del cliente-'+ tipo.value)
    f2.update_layout(title_text= 'Gasto: '+ tipo.value)

def respuesta(change):
    filtrar()

tipo.observe(respuesta, names = 'value')  
filtrar()
parte_inferior = widgets.HBox([tipo])   

visualizacion = widgets.VBox(
              [parte_inferior,
              f1, f2]
            )

display(visualizacion)
```
